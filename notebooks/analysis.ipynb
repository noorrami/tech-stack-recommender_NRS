# Tech Stack Recommender - Exploratory Data Analysis
This notebook performs EDA on the raw_skills.csv dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sys
sys.path.append('..')
from src.preprocess import load_and_preprocess_data

df = load_and_preprocess_data('../data/raw_skills.csv')
df.head()
```

```python
# Skill count distribution
df['skill_count'] = df['skills_list'].apply(len)
print(f"Average skills per job: {df['skill_count'].mean():.2f}")
plt.figure(figsize=(10,4))
sns.histplot(df['skill_count'], bins=30, kde=True)
plt.title('Distribution of Number of Skills per Job')
plt.show()
```

```python
# Most frequent skills
all_skills = []
for skills in df['skills_list']:
    all_skills.extend(skills)
skill_freq = pd.Series(all_skills).value_counts().head(15)
plt.figure(figsize=(10,6))
sns.barplot(x=skill_freq.values, y=skill_freq.index)
plt.title('Top 15 Most Frequent Skills')
plt.xlabel('Frequency')
plt.show()
```
